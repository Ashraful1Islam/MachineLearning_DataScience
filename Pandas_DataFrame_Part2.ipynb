```python
# Import necessary libraries
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# Set seed so results are reproducible
np.random.seed(101)
```

```python
# Create DataFrame with random numbers (5 rows x 4 columns)
# Rows: Aâ€“E, Columns: W, X, Y, Z
df=pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df
```

```python
# Create a boolean DataFrame: True where values > 0, False otherwise
booldf=df > 0
booldf
```

```python
# Use boolean DataFrame to filter original df (keeps values > 0, others become NaN)
df[booldf]
```

```python
# Shortcut version (same result as above)
df[df>0]
```

```python
df
```

```python
# Check where column 'W' is greater than 0 (returns a Series of True/False)
df['W']>0
```

```python
# Show rows where column 'W' is greater than 0
df[df['W']>0]
```

```python

# Show rows where column 'Z' is less than 0
df[df['Z']<0]
```

```python
# Store filtered result (W > 0) in a new DataFrame
resultdf=df[df['W']>0]
resultdf['X'] # Access column 'X' from the filtered DataFrame
```

```python
# Filter rows where W > 0, then show only columns Y and X
df[df['W']>0][['Y','X']]
```

```python
# Store condition into a boolean Series
boolser=df['W']>0
result=df[boolser]
mycols=['Y','X']# Create a list of desired column names
result[mycols]# Use that list to select specific columns from filtered rows
```

```python
# Use multiple conditions: rows where W > 0 AND Y > 1
df[(df['W']>0) & (df['Y']>1)]
```

```python
# Reset the index (converts row labels Aâ€“E into a normal column and resets index
df.reset_index()
```

```python
# Create a new list of state names and split them into individual strings
newind='CA NY WY OR CO'.split()
newind
```

```python
# Add a new column called 'States' using the new list
df['States']=newind
df
```

```python
# Set 'States' column as the new index
df.set_index('States')
```
