```python
# Import necessary libraries
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
#index Levels
# Creating MultiIndex
# Define two levels of index: outside (Groups) and inside (Numbers)
outside=['G1','G1','G1','G2','G2','G2']
inside=[1,2,3,1,2,3]
# Combine both into a list of tuples
hier_index=list(zip(outside,inside))
# Convert list of tuples into a MultiIndex
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
# Create a DataFrame with this hierarchical index and two columns: A and B
df=pd.DataFrame(randn(6,2),hier_index,['A','B'])
df
```

```python
# Access all rows under group 'G1'
df.loc['G1']
```

```python
# Access specific row: group 'G1', number 1
df.loc['G1'].loc[1]
```

```python

# Set names for the index levels for better readability
df.index.names=['Groups','Num']
df
```

```python

# Access a specific value: Group 'G2', Number 2, Column 'B'
df.loc['G2'].loc[2]['B']
```

```python
df.loc['G1'].loc[2]['B']
```

```python
# Use .xs to get all rows where Num = 1, across all groups
df.xs(1,level='Num')
```
